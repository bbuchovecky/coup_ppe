```python
kind = "a1b"
ensemble = "hadcm3"
rootpath = xppe.access.config.get_data_root_path(kind, ensemble)
outpath = "."

xppe.catalog.catalog.validate_ensemble_kind(ensemble, kind)
if ensemble == "hadcm3":
    file_format = "timeseries"
    depth = 0
else:
    xppe.catalog.catalog.validate_file_format(file_format)

# Get configuration and parser
MODELS = xppe.access.config.get_models()
PARSERS_CONFIG = xppe.access.config.get_parsers_config()
cfg = PARSERS_CONFIG[MODELS[ensemble]][file_format].copy()
if depth is not None:
    cfg["depth"] = depth

parser = xppe.catalog.catalog._get_parser(ensemble, file_format)
paths = xppe.catalog.catalog._normalize_paths(rootpath)

# Validate output path
outpath = Path(outpath).resolve()
if not outpath.exists():
    raise FileNotFoundError(f"Output directory does not exist: {outpath}")

# Create the catalog builder
builder = xppe.catalog.catalog._create_catalog_builder(paths, cfg, parser)
builder.clean_dataframe()
```

```python
catalog_name = f"{kind}_{ensemble}_{file_format}_catalog"
builder.save(
    name=catalog_name,
    directory=str(outpath),
    path_column_name="path",
    variable_column_name="variable",
    data_format="netcdf",
    groupby_attrs=cfg["groupby_attrs"],
    aggregations=cfg["aggregation"],
)
```
